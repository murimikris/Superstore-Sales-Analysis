<h5>What is the overall sales trend?

```python
#Get month and year from the order_date column
df["month-year"] = df["order_date"].apply(lambda x: x.strftime("%Y-%m"))
```

```python
trend=df.groupby("month-year").sum()["sales"].reset_index()
```

```python
plt.figure(figsize=(15,6))
plt.plot(trend["month-year"],trend["sales"])
plt.xticks(rotation="vertical")
plt.show()
```

### Sales and Profits in the last year

#### Which products had the highest sales in 2014?
          How does this compare to the overall sales for the whole dataset?

```python
# Filter data to have only 2014 data.

df_2014 = df[df['year'] == 2014]
df_2014.head()
```

```python
product_sales_2014 = pd.DataFrame(df_2014.groupby("product_name").sum()["sales"])
product_sales_2014.sort_values("sales", ascending=False, inplace=True)
#Top 10 products by sales
product_sales_2014.head(10)
```

```python
# Which were the most profitable products in 2014?
profit_2014 = pd.DataFrame(df_2014.groupby('product_name').sum()['profit'])
profit_2014.sort_values('profit', inplace=True, ascending=False)
profit_2014.head(10)
```

```python
# Which products made a loss in 2014 ?

loss_products = profit_2014[profit_2014['profit'] < 0]
loss_products
```

```python
# Which categories were the most profitable in the last year?

cat_profit = pd.DataFrame(df_2014.groupby('category').sum()['profit'])
cat_profit.sort_values('profit', ascending=False)
```

```python
# Which sub_categories were the most profitable in the last year?
sub_profit = pd.DataFrame(df_2014.groupby('sub_category').sum()['profit'])
sub_profit.sort_values('profit',ascending=False)
```

```python
# Which market is the most profitable?
market_profit = pd.DataFrame(df_2014.groupby('market').sum()[['profit', 'sales']])
market_profit.sort_values('profit', ascending=False)
```

```python
# Which countries are making losses?
profit_country = pd.DataFrame(df_2014.groupby('country').sum()['profit'])
loss_country = pd.DataFrame(profit_country[profit_country['profit']< 0])
loss_country.sort_values('profit', ascending=True)
```

### SHIPPING COSTS

```python
# Which market has the highest average shipping cost?

avg_ship_cost = pd.DataFrame(df_2014.groupby('market').mean()['shipping_cost'])
avg_ship_cost.sort_values('shipping_cost', ascending=False)
```

```python
# Shipping cost by category
cat_ship_cost = pd.DataFrame(df_2014.groupby(['category','sub_category']).mean()['shipping_cost'])
cat_ship_cost.sort_values('shipping_cost', ascending=False)
```

```python
# Shipping cost by country
avg_country_ship_cost = pd.DataFrame(df_2014.groupby('country').mean()['shipping_cost'])
avg_country_ship_cost.sort_values('shipping_cost', ascending=False, inplace=True)

# Most expensive countries to ship to?
avg_country_ship_cost.head(10)
```
